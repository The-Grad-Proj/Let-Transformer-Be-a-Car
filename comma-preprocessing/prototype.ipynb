```python
# Make sure the image is displayed correctly
img = cv2.imread(rf"{sample_folder}\preview.png")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for correct color display

 
```

```python
# Folder to save frames
os.makedirs(rf"{sample_folder}\frames", exist_ok=True)

# Convert video to frames
vidcap = cv2.VideoCapture(rf"{sample_folder}\video.hevc")
success, image = vidcap.read()
count = 0
while success:
    cv2.imwrite(rf"{sample_folder}\frames\frame_{count}.png", image)  # save frame as PNG file
    success, image = vidcap.read()
    count += 1

print(f"Total frames: {count}")
```

```python

# Display a random frame
img = cv2.imread(rf"{sample_folder}\frames\frame_100.png")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
